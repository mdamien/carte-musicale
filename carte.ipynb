# Cartographie automatique d'artistes

Les données sont dans `distances.json`, chaque artiste à une distance pouvant aller de 0 à 1 avec les autres artistes

```python
# Chargement des données

import json

distances = json.load(open('distances.json'))

etiquettes = list(sorted(distances.keys()))
print(etiquettes)

données = []
for i, artiste in enumerate(etiquettes):
    ligne = [i]
    for autre_artiste in etiquettes:
        if artiste == autre_artiste:
            ligne.append(0)
        else:
            dist = distances[artiste].get(autre_artiste)
            if dist is None:
                dist = distances[autre_artiste].get(artiste)
            if dist is None:
                raise Exception(f'Pas de distance entre {artiste} et {autre_artiste}')
            ligne.append(dist)
    données.append(ligne)
print(données)
```

```python
import umap
import umap.plot

def metric(a, b):
    dist = b[int(a[0])+1]
    # print(etiquettes[int(a[0])],'/',etiquettes[int(b[0])], '=', dist)
    return dist

mapper = umap.UMAP(metric=metric, random_state=30).fit(données)

umap.plot.output_notebook()
p = umap.plot.interactive(mapper, labels=etiquettes, hover_data={"index": etiquettes})
umap.plot.show(p)
```
